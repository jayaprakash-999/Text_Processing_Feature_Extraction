```python
import pandas as pd
book=pd.read_csv("apple.txt",sep='\t')

# Text file
# seperate ='tab space' any use
```

```python
book

# click at table u see text
```

```python
book = [x.strip() for x in book.x] # remove both the leading and the trailing characters
book = [x for x in book if x] # removes empty strings, because they are considered in Python as False
book[0:10]


# strip is used to remove extra space
# strip is used to all text in one particular corpus
#
# , means next value
# all data are single string
```

### Joining the list

```python
# Joining the list into one string/text
text = ' '.join(book)
text

## Entire data into a single cotation text
```

```python
len(text)
```

### #Punctuation

```python
#Punctuation

no_punc_text = text.translate(str.maketrans('', '', string.punctuation)) #with arguments (x, y, z) where 'x' and 'y'
# must be equal-length strings and characters in 'x'
# are replaced by characters in 'y'. 'z'
# is a string (string.punctuation here)

no_punc_text

##  remove punctuation ( , . - )
```

### #Tokenization

```python
#Tokenization
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
print(text_tokens[0:50])

# Tokenization --> unique word , no punchuvations
# Display first 50 characters
```

```python
len(text_tokens)
```

### Stopwords

```python
#Stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('punkt_tab')

# punkt_tab Direct download
```

```python
#Remove stopwords
import nltk
nltk.download('stopwords')

my_stop_words = stopwords.words('english')
my_stop_words.append('the')
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:40])


# This line adds the word "the" to our custom stopwords list.
# Although "the" is already in the default nltk stopwords list, this demonstrates(‡∞á‡∞¶‡∞ø ‡∞ö‡±Ç‡∞™‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø) how you can customize your stopwords list if needed
```

### lower_words

```python
#Noramalize the data
lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:25])

# all these words are converting into lower words
```

### Stemming

```python
#Stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens [0:40])


# stem is used to word to convert root
# ps.stem(word) edhi --> word in lower_words
# steamming (its, it,iting) --> it
```

### lemmatization

```python
#!python -m spacy download en
```

```python
# NLP english language model of spacy library
nlp = spacy.load('en_core_web_md')
```

```python
# lemmas being one of them, but mostly POS, which will follow later
doc = nlp(' '.join(no_stop_tokens))
print(doc[0:40])
```

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas[0:25])

# lemmatization not apply only list , tokenization
# lemmatization apply sentence only
# Out put == Coders --> coder , Got --> get
# All synonomes in on eword
```

# Feature Extraction

### CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(lemmas)
X
```

```python
# Use get_feature_names_out() instead of get_feature_names()

print(vectorizer.get_feature_names_out()[40:90])

# get_feature_names_out--> '4th' '50'
```
